# Лабораторная работа 2.1
Проверка теоремы Котельникова на сигнале $x(t) = sin(w_0 \cdot t) $

```python
w0 = 3
f = w0/(2*math.pi)
FD = f /2 
r = 5
fig1 = plt.figure(1)
axes1 = fig1.subplots(1, 1)
sin_sig = array([sin(w0 * t * 0.005) for t in range(int(r/0.005))])#график сигнала
axes1.plot(array([t * 0.005 for t in range(int(r/0.005))]), sin_sig, 'r')
axes1.grid(True)
plt.show()
```

```python
s = array([sin(w0 * t * FD) for t in range(int(r/FD))])
```

```python
y = []
cc = 0.005
for i in range(int(r/cc)):
    ssum = 0
    for j in range(int(r/FD)):
        if i*cc == j*FD:
            ssum+= s[j]
        else:
            ssum+= s[j] * math.sin(math.pi/FD * (i*cc  - j*FD  )) / (math.pi/FD * (i*cc - j*FD))
    y.append(ssum)
```

```python
print(s)
print(y)
```
